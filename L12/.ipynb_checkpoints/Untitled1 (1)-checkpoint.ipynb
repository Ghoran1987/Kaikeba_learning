```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
#一、加载训练集和测试集数据
file_name_train='used_car_train_20200313.csv'
file_name_test='used_car_testB_20200421.csv'
train_data= pd.read_csv(file_name_train, sep=' ')
test_data= pd.read_csv(file_name_test, sep=' ')
print(train_data.shape)
print(test_data.shape)
print('训练集的列名：',train_data.columns)
print('测试集的列名：',test_data.columns)
```

```python
#二、合并训练集和测试集并进行数据探索
concat_data=pd.concat([train_data,test_data])
```

```python
#三、进行数据预处理。1:notRapairedDamage中的'-’异常值处理
train_data['notRepairedDamage']=train_data['notRepairedDamage'].replace('-',0).astype('float16')
test_data['notRepairedDamage']=test_data['notRepairedDamage'].replace('-',0).astype('float16')
train_data['notRepairedDamage'].value_counts()
```

```python
#三、进行数据预处理。2:使用众数将数据集中的缺失值填充完整
train_data=train_data.fillna(train_data.mode().iloc[0,:])
train_data.isnull().sum()
```

```python
#三、进行数据预处理。3：通过热力图，分析出与price相关性较大的特征，用于训练
plt.rcParams['figure.figsize']=(15,10)#设置图幅大小
corrmatrix=train_data.corr()#计算相关系数
sns.heatmap(corrmatrix,square=True,vmax=1,vmin=-1,center=0.0,cmap='coolwarm')#绘制热力图
plt.show()
```

```python
#三、进行数据预处理。3：输出和price的相关性排名：
co=train_data.corr()[u'price']
print(co.sort_values(ascending=False))
```

```python
#三、进行数据预处理。4：从相关性分析可以看出creatDate、name、SaleID、seller与price相关性极小，这里就不做考虑，直接从数据集中去除。
train_X = train_data.drop(['SaleID','creatDate', 'name','price','seller'], axis=1)
train_y = train_data['price']
test_data = test_data.drop(['SaleID','creatDate', 'name','seller'], axis=1)
print(train_X)
print(train_y)
```

```python
#三、进行数据预处理。5：数据归一化
scaler=MinMaxScaler(feature_range=(0, 1))
train_newX= scaler.fit_transform(train_X)

test_newX=scaler.transform(test_data)
print(train_newX)
print(test_newX)
```

```python
#四、采用XgBoost模型进行训练。
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

#自定义MAE评估函数
def myFeval(preds,train_data):
    label = train_data.get_label()
    score=mean_absolute_error(label,preds)
    return 'myFeval',score

param = {                #'boosting_type':'gbdt',
                         'objective' : 'reg:linear', 
                         'eta' : 0.05,
                         'max_depth' : 11,
                         'colsample_bytree':0.8,
                         'subsample': 0.9,
                         #'subsample_freq': 8,
                         'alpha': 0,
                         'lambda':1,
                         'gamma':1,
                         'min_child_weight':1,
                         #'n_estimators':250,
                         'seed':2000
        }
X_train, X_valid, y_train, y_valid = train_test_split(train_newX, train_y, test_size=0.2, random_state=42)

train_data = xgb.DMatrix(X_train, label=y_train)
valid_data = xgb.DMatrix(X_valid, label=y_valid)
test_data = xgb.DMatrix(test_newX)

model = xgb.train(param, train_data, evals=[(train_data, 'train'), (valid_data, 'valid')], num_boost_round = 10000, early_stopping_rounds=50, verbose_eval=25,feval = myFeval)
```

```python
predict = model.predict(test_data)
sub_test= pd.read_csv(file_name_test, sep=' ')[['SaleID']].copy()
sub_test['price']=predict
print(sub_test)
sub_test[['SaleID','Attrition']].to_csv('predictions.csv')
```
