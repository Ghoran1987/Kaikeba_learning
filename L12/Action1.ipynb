```python
import networkx as nx
from networkx import to_numpy_matrix
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.read_gml('dolphins.gml')
```

```python
# 网络G进行可视化
def plot_graph(G):
    plt.figure()
    pos = nx.spring_layout(G)
    edges = G.edges()
    nx.draw_networkx(G, pos, edges=edges);
    nx.draw_networkx_nodes(G, pos, nodelist=G.nodes(), node_size=300, node_color='r', alpha=0.8)
    nx.draw_networkx_edges(G, pos, edgelist=edges,alpha =0.4)
    plt.show()
```

```python
plot_graph(G)
print(list(G.nodes()))
```

```python
# 一、构建GCN:1.按照字母顺序排序
order = sorted(list(G.nodes()))
print(order)
```

```python
# 一、构建GCN:2.生成邻接矩阵
A = to_numpy_matrix(G, nodelist=order)
print('A=\n', A) 
```

```python
# 一、构建GCN: 3.生成对角矩阵
I = np.eye(G.number_of_nodes())
```

```python
# 一、构建GCN:4.给邻接矩阵A加上一个单位矩阵I，生成A_hat
A_hat = A + I
print('A_hat=\n', A_hat)
```

```python
# 一、构建GCN:5. 生成A_hat的度矩阵D_hat
D_hat = np.array(np.sum(A_hat, axis=0))[0]
print('D_hat=\n', D_hat)
```

```python
# 一、构建GCN:6.得到D_hat对角线上的元素
D_hat = np.matrix(np.diag(D_hat))
print('D_hat=\n', D_hat)
```

```python
# 一、构建GCN:7.初始化权重。 normal 正态分布 loc均值 scale标准差
W_1 = np.random.normal(loc=0, scale=1, size=(G.number_of_nodes(), 4))
W_2 = np.random.normal(loc=0, size=(W_1.shape[1], 2))
print('W_1=\n', W_1)
print('W_2=\n', W_2)
```

```python
# 一、构建GCN:8_1. 采用relu()激活函数
#x<0时 结果=0; x>=0时，结果=x
def relu(x):
    return(abs(x)+x)/2
# 叠加GCN层，这里只使用单位矩阵作为特征表征，即每个节点被表示为一个 one-hot 编码的类别变量
def gcn_layer_relu(A_hat, D_hat, X, W):
    return relu(D_hat**-1 * A_hat * X * W)

H_1_relu = gcn_layer_relu(A_hat, D_hat, I, W_1)
H_2_relu = gcn_layer_relu(A_hat, D_hat, H_1_relu, W_2)
output_relu = H_2_relu
print('output with relu=\n', output_relu)
```

```python
# 一、构建GCN:8_2. 不采用relu()激活函数
def gcn_layer(A_hat, D_hat, X, W):
    return D_hat**-1 * A_hat * X * W
H_1 = gcn_layer(A_hat, D_hat, I, W_1)
H_2 = gcn_layer(A_hat, D_hat, H_1, W_2)
output= H_2
print('output without relu=\n', output)
```

```python
# 二、提取特征表征
def get_feature(output):
    feature_representations = {}
    nodes = list(G.nodes())
    for i in range(len(nodes)):
        feature_representations[nodes[i]] = np.array(output)[i]
    '''print('feature_representations=\n', feature_representations)'''
    return feature_representations
```

```python
#二、提取特征表征。1.relu表征特征
get_feature(output_relu)
```

```python
#二、提取特征表征。2.无relu表征特征
get_feature(output)
```

```python
# 三、绘制output，节点GCN embedding可视化
def plot_node(output,title):
    nodes = list(G.nodes())
    for i in range(len(nodes)):
        node_name = nodes[i]
        plt.scatter(np.array(output)[i,0],np.array(output)[i,1] ,label=str(i),color ='green',alpha=0.5,s = 250)
        plt.text(np.array(output)[i,0],np.array(output)[i,1] ,i, horizontalalignment='center',verticalalignment='center', fontdict={'color':'black'})
    plt.title(title)
    plt.show()
```

```python
# 三、绘制output。1.绘制Graph Embedding With relu
plot_node(output_relu,'Graph Embedding With relu')
```

```python
# 三、绘制output。2.绘制Graph Embedding Without relu
plot_node(output,'Graph Embedding Without relu')
```
