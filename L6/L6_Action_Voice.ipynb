```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
#获取数据
data_voice=pd.read_csv("voice.csv")
```

```python
print(data_voice)
```

```python
#分离特征sparse_features和目标targets
data=pd.DataFrame(data_voice)
X = data.iloc[:,:-1]
y =data.iloc[:,-1]
print(X)
```

```python
#target_y=np.array([1 if x=='male' else 0 for x in target]).reshape(-1,1)
#print(target_y)

encoder=LabelEncoder()
y=encoder.fit_transform(y)
```

```python
# 对特征标签进行编码
scaler = StandardScaler()
scaler.fit(X)
X=scaler.transform(X)
#data=np.hstack((trans_data,target_y))
print(X)
print(X.shape)
```

```python
# 将数据集切分成训练集和测试集
from sklearn.model_selection import train_test_split
X_train,X_valid, y_train, y_valid = train_test_split(X,y, test_size=.2)

print(X_train.shape)
print(y_train.shape)
print(X_valid.shape)
#print(y_train)
#train_model_input = {name:train[name].values for name in feature_names}
#test_model_input = {name:test[name].values for name in feature_names}
#print(train_model_input)
```

```python
# 使用XGBoost工具对男女声音进行识别
import xgboost as xgb
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score

model = xgb.XGBClassifier()
model.fit(X_train,y_train)
prediction = model.predict(X_valid)
print('Accuracy', accuracy_score(y_valid, prediction))
```

```python

from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
prediction = svc.predict(X_valid)
print('Accuracy', accuracy_score(y_valid, prediction))
```
